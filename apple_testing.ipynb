```python
import requests
import pandas as pd
```

```python
# get company specific filings metadata
headers = {'User-Agent': "jose.trindade@bts.tech"}
aapl_company_facts = requests.get(
    f'https://data.sec.gov/api/xbrl/companyfacts/CIK0000320193.json',
    headers=headers)

aapl_company_facts_json = aapl_company_facts.json()
```

```python
json_data = aapl_company_facts_json["facts"]["us-gaap"]
```

```python
#json_data
```

```python
data_list = []

for key, value in json_data.items():
    for unit, records in value['units'].items():
        for record in records:
            data_list.append({
                'label': value['label'],
                'end': record['end'],
                'value': record['val'],
                'account number': record['accn'],
                'fy': record['fy'],
                'fp': record['fp'],
                'form': record['form'],
                'filed date': record['filed']
            })

df = pd.DataFrame(data_list)
```
