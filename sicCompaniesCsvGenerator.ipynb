```python
import pandas as pd
```

```python
companies_per_sic = pd.read_csv("companies_per_sic.csv", index_col=0).sort_values("count", ascending=False)
top_10 = companies_per_sic[:10] # TOP 10
top_10
```

```python
company_data_df = pd.read_csv("Company_Data.csv", index_col=0)
company_data_df
```

```python
for ind, sic in enumerate(top_10.sic.tolist()):
    # Name for the file
    name = str(top_10.sicDescription[ind]) + ".csv"
    df = company_data_df[company_data_df["sic"]==sic][["cik_str","ticker","title","sic","sicDescription"]]
    df.to_csv(name)
```
